```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame({
    'Age':[30,40,50,60,40],
    'Income':[200,300,800,600,300],
    'Yrs worked':[10,20,20,20,20],
    'Vacation':[4,4,1,2,5]
})
df.head()
```

### Population

```python
sns.heatmap(np.cov(df.to_numpy().T,bias=True),annot=True)
```

### Sample

```python
sns.heatmap(np.cov(df.to_numpy().T,bias=False),annot=True)
```

```python
sns.heatmap(pd.DataFrame.cov(df),annot=True)
```
