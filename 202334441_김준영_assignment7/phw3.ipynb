```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'spends':[2400,2650,2350,4950,3100,2500,5106,3100,2900,1750],
    'income':[41200,50100,52000,66000,44500,37700,73500,37500,56700,35600]
})
df.head()
```

```python
model = linear_model.LinearRegression()
model.fit(df['spends'].to_numpy()[:,np.newaxis],df['income'].to_numpy()[:,np.newaxis])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
sns.scatterplot(df,x='spends',y='income')
x= np.arange(np.min(df['spends']),np.max(df['spends']),0.1)
plt.plot(x,(x*model.coef_ + model.intercept_)[0],color='r')
plt.show()
```
